```python
bb["Close"]
```

```python
def rank_pref(price):
    if price <= 10:
        return "Poor"
    elif price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
bb["Close"].apply(rank_pref)
```

```python
bb["Close"].apply(rank_pref).value_counts()
```

```python
bb["Close"].apply(rank_pref).value_counts().plot()
```

```python
bb["Close"].apply(rank_pref).value_counts().plot(kind = "line")
```

```python
bb["Close"].apply(rank_pref).value_counts().plot(kind = "bar")
```

```python
bb["Close"].apply(rank_pref).value_counts().plot(kind = "barh")
```

# Creating Pie Charts to Represent Proportions

```python
bb["Close"].mean()
```

```python
def rank_perf(price):
    if price >= 29.452764073094578:
        return "Above Average"
    else:
        return "Below Average"
bb["Close"].apply(rank_perf)
```

```python
plt.style.use('dark_background')
```

```python
bb["Close"].apply(rank_perf).value_counts().plot(kind = "pie", legend = True)
```
