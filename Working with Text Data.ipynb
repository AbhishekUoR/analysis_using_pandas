```python
import pandas as pd
```

```python
chicago = pd.read_csv("chicago.csv")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head()
```

# Common string methods title(), .lower(), .upper()

```python
chicago["Name"].str.title()
```

```python
chicago["Name"] = chicago["Name"].str.title()
chicago.head()
```

# replace()

```python
chicago["Department"].str.replace("MGMNT", "Management")
```

```python
chicago["Department"] = chicago["Department"].str.replace("MGMNT", "Management")
```

```python
chicago.head()
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$", "")
```

```python
chicago["Employee Annual Salary"].astype("float")
```

```python
chicago["Employee Annual Salary"].mean
```

# filtering with string method

```python
chicago.dropna(how = "all", inplace = True)
chicago.tail(3)
```

```python
chicago["Position Title"].str.lower().str.contains("water")
```

```python
chicago[chicago["Position Title"].str.lower().str.contains("water")]
```

```python
chicago[chicago["Position Title"].str.lower().str.startswith("water")]
```

```python
chicago[chicago["Position Title"].str.lower().str.endswith("ist")]
```
