# extract rows in multi index using loc[] and iloc[]

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.loc[("2016-01-01", "Argentina"), "Price in US Dollars"]
```

```python
bigmac.loc[("2016-01-01", ), "Price in US Dollars"]
```

```python
bigmac.iloc[0, 0]
```

```python
bigmac.iloc[0]
```

```python
bigmac.iloc[0:4]
```

```python
bigmac.iloc[[0, 4]]
```

# transpose

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"] ,index_col = ["Date", "Country"]).sort_index()
bigmac.head()
```

```python
bigmacT = bigmac.transpose()
bigmacT.head()
```

```python
bigmacT.loc["Price in US Dollars"]
```

```python
bigmacT.loc["Price in US Dollars",]
```

```python
bigmacT.head(3)
```

```python
bigmacT.loc["Price in US Dollars", "2010-01-01"]
```

```python
bigmacT.loc["Price in US Dollars", ("2010-01-01", "China")]
```

# pivot()

```python
salesman = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
salesman["Salesman"] = salesman["Salesman"].astype("category")
```

```python
salesman.head(3)
```

```python
salesman.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

```python
salesman2 = salesman.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```
